```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = [310, 400, 800]

model_dict = {
    'neural': 'NODE',
    'expert': 'Expert',
    'residual': 'Residual',
    'ensemble': 'Ensemble',
    'hybrid': 'LHM'
}
```

```python

def create_plot(s):
    for n in model_dict.keys():
        result_path="../results/sample_{}_{}_{}.pkl".format(s, n, 12)
        with open(result_path, 'rb') as f:
            res = pickle.load(f)
        f = 1
        if n == 'hybrid':
            plt.fill_between(np.arange(3) + 1, res['rmse_x'] - f * res['rmse_x_sd'], res['rmse_x'] + f * res['rmse_x_sd'], alpha=0.3, color=colors[5])
            plt.plot(np.arange(3) + 1, res['rmse_x'], 'o-', label=model_dict[n], color=colors[5])
        else:
            plt.fill_between(np.arange(3) + 1, res['rmse_x'] - f * res['rmse_x_sd'], res['rmse_x'] + f * res['rmse_x_sd'], alpha=0.3)
            plt.plot(np.arange(3) + 1, res['rmse_x'], 'o-', label=model_dict[n])
        ax = plt.gca()
        ax.set_xticks(np.arange(3) + 1)
        ax.set_xticklabels(np.arange(3) + 1)
    plt.ylim(0.4, 0.9)
    if s == 400:
        plt.title('RMSE\n Sample Size N={}'.format(s-300))
    else:
        plt.title('Sample Size N={}'.format(s-300))
    plt.ylabel('RMSE')
    plt.xlabel('k step head')
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
plt.style.use('tableau-colorblind10')
plt.rcParams['font.size'] = '15'

plt.figure(figsize=(14,2.5))

plt.subplot(131)
create_plot(310)

plt.subplot(132)
create_plot(400)

plt.subplot(133)
create_plot(800)

plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.tight_layout(pad=0.1)

plt.savefig(fname='Fig7.png', dpi=300)
```
