```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('results/results_z.txt', header = None)
df = df.sort_values(1)
df.head()
df_rmse = df[df[2] == 'rmse_x']
df_crps = df[df[2] == 'cprs_x']
```

```python
plt.figure(figsize=(7,2.5))
# plt.figure(figsize=(14,2.5))
plt.style.use('seaborn-colorblind')
plt.rcParams['font.size'] = '13'

# RMSE-X-Sample
plt.subplot(121)
plt.bar(df_rmse[1], df_rmse[3], yerr=df_rmse[4], width=3)

plt.title('RMSE')
ax = plt.gca()
ax.set_xticks([6, 10, 15])
ax.set_xticklabels([6, 10, 15])
plt.xlabel('Number of latent variables Z')

# CRPS-X-Sample
plt.subplot(122)
plt.bar(df_crps[1], df_crps[3], yerr=df_crps[4], width=3)

plt.title('CRPS')
plt.xlabel('Number of latent variables Z')
ax = plt.gca()
ax.set_xticks([6, 10, 15])
ax.set_xticklabels([6, 10, 15])

plt.tight_layout(pad=0.)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.7)

plt.savefig(fname='Fig9.png', dpi=300)
```

```python
df_flow = pd.read_csv('results/exp_flow.csv', header = None)
df_lhm = pd.read_csv('results/exp_lhm.csv', header = None)
```

```python
plt.figure(figsize=(7,2.5))
# plt.figure(figsize=(14,2.5))
plt.style.use('tableau-colorblind10')
plt.rcParams['font.size'] = '13'

# RMSE-X-Sample
plt.subplot(121)
plt.plot(df_flow[0], df_flow[2], label='LHM-NF')
plt.plot(df_lhm[0], df_lhm[2], label='LHM')

plt.title('Loss (Neg. ELBO) - Training')
plt.xlabel('Training iterations')

# CRPS-X-Sample
plt.subplot(122)
plt.plot(df_flow[0], df_flow[1], label='LHM-NF')
plt.plot(df_lhm[0], df_lhm[1], label='LHM')

plt.title('Loss (Neg. ELBO) - Evaluation')
plt.xlabel('Training iterations')

plt.legend(loc="upper right")
plt.tight_layout(pad=0.)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

plt.savefig(fname='Fig10.png', dpi=300)
```
