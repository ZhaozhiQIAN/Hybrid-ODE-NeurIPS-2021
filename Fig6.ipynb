```python
import pandas as pds
import matplotlib.pyplot as plt
import numpy as np
```

```python
t0 = 5

if t0 == 5:
    t = ''
else:
    t = '_' + str(t0)
```

## sample

```python
def get_sample(t0=2):

    if t0 == 5:
        t = ''
    else:
        t = '_' + str(t0)
    df = pds.read_csv('results/results_sample{}.txt'.format(t), header = None)

    df_rmse = df[df[2] == 'rmse_x']
    df_cprs = df[df[2] == 'cprs_x']

    df_rmseh = df[df[2] == 'rmse_z0']
    df_cprsh = df[df[2] == 'cprs_z0']

    ## dimension

    df2 = pds.read_csv('results/results_dim{}.txt'.format(t), header = None)

    df2 = df2.sort_values([0,1,2])
    df_rmse2 = df2[df2[2] == 'rmse_x']
    df_cprs2 = df2[df2[2] == 'cprs_x']

    df_rmseh2 = df2[df2[2] == 'rmse_z0']
    df_cprsh2 = df2[df2[2] == 'cprs_z0']
    return df_rmse, df_cprs, df_rmse2, df_cprs2
    
```

```python
f=3
model_dict = {
    'neural': 'NODE',
    'expert': 'Expert',
    'residual': 'Residual',
    'ensemble2': 'Ensemble',
    'hybrid': 'LHM'
}

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

plt.figure(figsize=(14,2.5))
plt.style.use('tableau-colorblind10')
plt.rcParams['font.size'] = '13'

# RMSE-X-Sample
plt.subplot(131)

df_rmse, df_cprs, df_rmse2, df_cprs2 = get_sample(5)

for n in model_dict.keys():
    dfs = df_rmse[df_rmse[0] == n]
    if n == 'hybrid':
        plt.fill_between(dfs[1] - 300, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3, color=colors[5])
        plt.plot(dfs[1] - 300, dfs[3], 'o-', label=model_dict[n], color=colors[5])
    else:
        plt.fill_between(dfs[1] - 300, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
        plt.plot(dfs[1] - 300, dfs[3], 'o-', label=model_dict[n])
        
plt.title(r'RMSE ($t_0=5$)')
plt.xscale('log')
plt.ylim(0.5, 1.1)
ax = plt.gca()
ax.set_xticks([10,100,1000])
ax.set_xticklabels([10,100,1000])
plt.xlabel('Sample size N')


# RMSE-X-Sample
plt.subplot(132)

df_rmse, df_cprs, df_rmse2, df_cprs2 = get_sample(10)

for n in model_dict.keys():
    dfs = df_rmse[df_rmse[0] == n]
    if n == 'hybrid':
        plt.fill_between(dfs[1] - 300, dfs[3] / f5 - f * dfs[4] / f5, dfs[3] / f5 + f * dfs[4] / f5, alpha=0.3, color=colors[5])
        plt.plot(dfs[1] - 300, dfs[3] / f5, 'o-', label=model_dict[n], color=colors[5])
    else:
        plt.fill_between(dfs[1] - 300, dfs[3] / f5 - f * dfs[4] / f5, dfs[3] / f5 + f * dfs[4] / f5, alpha=0.3)
        plt.plot(dfs[1] - 300, dfs[3] / f5, 'o-', label=model_dict[n])
        
plt.title(r'RMSE ($t_0=10$)')
plt.xscale('log')
plt.ylim(0.5, 1.1)
ax = plt.gca()
ax.set_xticks([10,100,1000])
ax.set_xticklabels([10,100,1000])
plt.xlabel('Sample size N')


# RMSE-X-Sample
plt.subplot(133)

df_rmse, df_cprs, df_rmse2, df_cprs2 = get_sample(12)

for n in model_dict.keys():
    dfs = df_rmse[df_rmse[0] == n]
    if n == 'hybrid':
        plt.fill_between(dfs[1] - 300, (dfs[3] - f * dfs[4])/ f10, (dfs[3] + f * dfs[4])/ f10, alpha=0.3, color=colors[5])
        plt.plot(dfs[1] - 300, dfs[3]/ f10, 'o-', label=model_dict[n], color=colors[5])
    else:
        plt.fill_between(dfs[1] - 300, (dfs[3] - f * dfs[4])/ f10, (dfs[3] + f * dfs[4])/ f10, alpha=0.3)
        plt.plot(dfs[1] - 300, dfs[3]/ f10, 'o-', label=model_dict[n])
        
plt.title(r'RMSE ($t_0=12$)')
plt.xscale('log')
plt.ylim(0.5, 1.1)
ax = plt.gca()
ax.set_xticks([10,100,1000])
ax.set_xticklabels([10,100,1000])
plt.xlabel('Sample size N')


# RMSE-X-Dim
# plt.subplot(143)
# for n in model_dict.keys():
#     dfs = df_rmse2[df_rmse2[0] == n]
    
#     if n == 'hybrid':
#         plt.fill_between(dfs[1] - 4, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3, color=colors[5])
#         plt.plot(dfs[1] - 4, dfs[3], 'o-', label=model_dict[n], color=colors[5])
#     else:
#         plt.fill_between(dfs[1] - 4, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
#         plt.plot(dfs[1] - 4, dfs[3], 'o-', label=model_dict[n])
# plt.title('RMSE')
# plt.ylim(0.6, 1.1)
# plt.xscale('log')
# ax = plt.gca()
# ax.xaxis.set_minor_formatter(plt.NullFormatter())
# ax.set_xticks([2, 4, 8])
# ax.set_xticklabels([2, 4, 8])
# plt.xlabel('Latent dimension M')

plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.tight_layout(pad=0.1)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.7)

# plt.savefig(fname='sim_plot_time.png', dpi=300)
```

```python
plt.figure(figsize=(12,5))
# plt.figure(figsize=(14,2.5))
plt.style.use('seaborn-colorblind')
plt.rcParams['font.size'] = '13'

# RMSE-X-Sample
plt.subplot(221)
for n in model_dict.keys():
    dfs = df_rmse[df_rmse[0] == n]
    plt.fill_between(dfs[1] - 300, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
    plt.plot(dfs[1] - 300, dfs[3], label=model_dict[n])
plt.title('RMSE')
plt.xscale('log')
ax = plt.gca()
ax.set_xticks([10,100,1000])
ax.set_xticklabels([10,100,1000])
plt.xlabel('Sample size N')

# CRPS-X-Sample
plt.subplot(222)
for n in model_dict.keys():
    dfs = df_cprs[df_cprs[0] == n]
    plt.fill_between(dfs[1] - 300, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
    plt.plot(dfs[1] - 300, dfs[3], label=model_dict[n])
plt.title('CRPS')
plt.xscale('log')
ax = plt.gca()
ax.set_xticks([10,100,1000])
ax.set_xticklabels([10,100,1000])
plt.xlabel('Sample size N')

# RMSE-X-Dim
plt.subplot(223)
for n in model_dict.keys():
    dfs = df_rmse2[df_rmse2[0] == n]
    plt.fill_between(dfs[1] - 4, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
    plt.plot(dfs[1] - 4, dfs[3], label=model_dict[n])
plt.title('RMSE')
plt.xscale('log')
ax = plt.gca()
ax.xaxis.set_minor_formatter(plt.NullFormatter())
ax.set_xticks([2, 4, 8])
ax.set_xticklabels([2, 4, 8])
plt.xlabel('Latent dimension M')

# CRPS-X-Dim
plt.subplot(224)
for n in model_dict.keys():
    dfs = df_cprs2[df_cprs2[0] == n]
    plt.fill_between(dfs[1] - 4, dfs[3] - f * dfs[4], dfs[3] + f * dfs[4], alpha=0.3)
    plt.plot(dfs[1] - 4, dfs[3], label=model_dict[n])
plt.title('CRPS')
plt.xscale('log')
plt.xlabel('Latent dimension M')

ax = plt.gca()
ax.xaxis.set_minor_formatter(plt.NullFormatter())
ax.set_xticks([2, 4, 8])
ax.set_xticklabels([2, 4, 8])

plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.tight_layout(pad=0.1)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.7)

plt.savefig(fname='sim_plot2.png')
```
